```python
import nltk

nltk.download('punkt')

nltk.download('stopwords')

nltk.download('vader_lexicon')
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
from wordcloud import WordCloud
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
def analyze_sentiment(sentence):
     # Tokenize the sentence into words
    words = nltk.word_tokenize(sentence.lower())
    
       # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    
     # Perform sentiment analysis
    analyzer = SentimentIntensityAnalyzer()
    sentiment_score = analyzer.polarity_scores(sentence)
    
    # Get the keywords and their counts
    keyword_counts = nltk.FreqDist(words)
    keywords = keyword_counts.keys()
    
     # Display the keywords and their counts
    print("Keywords and their counts:")
    for keyword in keywords:
        print(keyword + ": " + str(keyword_counts[keyword]))
        
      # Plot the word cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(keyword_counts)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
    
    
    # Return the sentiment score
    return sentiment_score
```
